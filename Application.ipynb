# Increase the quality of a document scanned
    Using thresholding methods to increase the quality of an image generated by a scanner.

```python
# Author: Muriki Gusmão Yamanaka

import cv2 as cv
import matplotlib
import numpy as np
import math
from matplotlib import pyplot as plt
from ipywidgets import interact, interactive, interact_manual

assembly = cv.imread('./Assets/assembly.jpg')
assembly = cv.cvtColor(assembly, cv.COLOR_BGR2GRAY)
essay = cv.imread('./Assets/sakubun.jpg')
essay = cv.cvtColor(essay, cv.COLOR_BGR2GRAY)
```

```python
imgs = ['assembly', 'essay']
# Sharpening kernel
sharp_kernel = np.array([[0, -1,  0],
                         [-1, 5, -1],
                         [0, -1,  0]])

def treatment (scale, img):
    fig, axes = plt.subplots(1, 2, figsize=(14, 12))
    
    if img == 'assembly':
        #output = cv.filter2D(src=assembly, ddepth=-1, kernel=sharp_kernel)   
        #ret,output = cv.threshold(output,scale,255,cv.THRESH_TOZERO)
        ret,output = cv.threshold(assembly,scale,255,cv.THRESH_TOZERO)
        axes[0].imshow(assembly, 'gray')
        axes[1].imshow(output, 'gray')
    elif img == 'essay':
        #output = cv.filter2D(src=essay, ddepth=-1, kernel=sharp_kernel)   
        #ret,output = cv.threshold(output,scale,255,cv.THRESH_TOZERO)
        ret,output = cv.threshold(essay,scale,255,cv.THRESH_TOZERO)
        axes[0].imshow(essay, 'gray')
        axes[1].imshow(output, 'gray')

interactive(treatment, scale=(150, 255), img=imgs)
```
