```python
import cv2 as cv
import matplotlib
import numpy as np
import math
from matplotlib import pyplot as plt

muriki = cv.imread('./Assets/muriki.png')
muriki_color = cv.cvtColor(muriki, cv.COLOR_BGR2RGB) #Change color to RGB
muriki_gray = cv.cvtColor(muriki, cv.COLOR_BGR2GRAY) #Change image to gray
```

# Image Plot

```python
plt.imshow(muriki_color, 'gray') #Plot Image
```

# Resize Image
## Alteração da resolução da imagem

```python
scale_percent = 800
width = int(muriki_color.shape[1] * scale_percent / 100)
height = int(muriki_color.shape[0] * scale_percent / 100)
dim = (width, height)

# resize image
muriki_resized = cv.resize(muriki_color, dim, interpolation = cv.INTER_AREA)
plt.imshow(muriki_resized, 'gray')
plt.imsave('Assets/raw_expand.png', muriki_resized)
print("Size of image after resize: ", muriki_resized.shape)
```

# UpSample
Why blurring? Because this reduces the aliasing or ringing effects that may arise if we downsample directly

```python
muriki_upsample = cv.pyrUp(muriki_color)

for i in range(2):
    muriki_upsample = cv.pyrUp(muriki_upsample)

plt.imshow(muriki_upsample, 'gray')
plt.imsave('Assets/raw_upsample.png', muriki_upsample)
print("Size of image after upsample: ", muriki_upsample.shape)
```
